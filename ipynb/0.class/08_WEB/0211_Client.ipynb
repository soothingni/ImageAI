# 2. Socket

## 2-2. 클라이언트

```python
import socket

sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

sock.connect(('localhost', 12345))    #바인딩은 아니고, 그냥 이 주소로 접속해달라는 뜻
print("서버접속 성공")

sock.send('hello'.encode())   #바이트 데이터로 바꿔줘야한다.
print("send message")

data = sock.recv(65535)
print("recieve: ", data.decode())
print("종료")
```
